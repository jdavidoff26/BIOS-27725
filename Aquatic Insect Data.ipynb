```python
def salinity_classification(salinity):
    """
    Classifies salinity levels into 'freshwater' or 'brackish'.
    
    Parameters:
    salinity (float): The salinity level to classify.
    
    Returns:
    str: The classification ('freshwater' or 'brackish').
    """
    if salinity < 1.01:
        return 'Freshwater'
    else:
        return 'Brackish'
```

```python
salinity_df['Salinity Classification'] = salinity_df['Salinity (S.G.)'].apply(salinity_classification)
salinity_df
```

```python
ai_extended_df['Salinity Classification'] = ai_extended_df['Salinity (S.G.)'].apply(salinity_classification)
ai_extended_df
```

```python
ai_chitest_df = ai_extended_df[['Species Name', 'Salinity Classification']]
ai_chitest_df
```

```python
import pandas as pd
from scipy.stats import fisher_exact
import numpy as np
from scipy.stats import norm

results_list = []

# Function to calculate the Wilson score interval
def wilson_score_interval(successes, trials, confidence=0.95):
    """
    Calculate the Wilson score interval for a proportion.
    
    Parameters:
    successes (int): Number of successes (e.g., observations in freshwater).
    trials (int): Total number of trials (e.g., total observations for the species).
    confidence (float): Confidence level for the interval.
    
    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    p = successes / trials
    z = norm.ppf((1 + confidence) / 2)
    denominator = 1 + z**2 / trials
    centre_adjusted_probability = p + z**2 / (2 * trials)
    adjusted_standard_deviation = np.sqrt((p * (1 - p) + z**2 / (4 * trials)) / trials)
    
    lower_bound = (centre_adjusted_probability - z * adjusted_standard_deviation) / denominator
    upper_bound = (centre_adjusted_probability + z * adjusted_standard_deviation) / denominator
    
    return lower_bound, upper_bound

def adjust_p_value(p_value, sample_size, threshold=0.05):
    """
    Adjust p-value based on sample size and use Bonferroni correction
    
    Parameters:
    p_value (float): The original p-value from Fisher's Exact Test.
    sample_size (int): The number of observations for the species.
    threshold (float): The base significance threshold.
    
    Returns:
    float: Adjusted p-value.
    """
    adjusted_p_value = p_value * 27
    return adjusted_p_value

# Loop through each species and perform Fisher's Exact Test
for species in ai_chitest_df['Species Name'].unique():
    # Filter data for the current species
    species_df = ai_chitest_df[ai_chitest_df['Species Name'] == species]
    
    # Count the number of freshwater and brackish observations
    freshwater_count = (species_df['Salinity Classification'] == 'Freshwater').sum()
    brackish_count = (species_df['Salinity Classification'] == 'Brackish').sum()
    total_count = len(species_df)
    
    # Create a contingency table
    contingency_table = pd.crosstab(species_df['Salinity Classification'], species_df['Species Name'])
    
    # Ensure contingency table is in 2x2 format
    if contingency_table.shape == (2, 2):
        table_array = contingency_table.values
    else:
        # Convert to 2x2 format if needed (e.g., when dealing with more than two categories)
        table_array = np.array([
            [freshwater_count, brackish_count],
            [total_count - freshwater_count, total_count - brackish_count]
        ])
    
    # Perform Fisher's Exact Test
    odds_ratio, p = fisher_exact(table_array)
    
    # Adjust the p-value based on sample size
    adjusted_p_value = adjust_p_value(p, total_count)
    
    # Calculate the confidence interval for the proportion of freshwater observations
    ci_lower, ci_upper = wilson_score_interval(freshwater_count, total_count)
    
    # Append the results
    results_list.append({
        'Species Name': species,
        'Odds Ratio': odds_ratio,
        'p-value': p,
        'Adjusted p-value': adjusted_p_value,
        'Freshwater Proportion': freshwater_count / total_count,
        'CI Lower Bound': ci_lower,
        'CI Upper Bound': ci_upper
    })

# Convert the list of results to a DataFrame
results = pd.DataFrame(results_list)

results
```

```python
plt.figure(figsize=(14, 8))
plt.barh(results['Species Name'], results['Adjusted p-value'], color='skyblue', edgecolor='black')
plt.axvline(x=0.05, color='red', linestyle='--', label='Significance Threshold (0.05)')
plt.xlabel('Adjusted p-value', fontsize=14)
plt.ylabel('Species Name', fontsize=14)
plt.title('Adjusted p-values for Species Distribution', fontsize=16)
plt.legend()
plt.tight_layout()
plt.show()
```
